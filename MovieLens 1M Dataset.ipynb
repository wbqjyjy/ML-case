```python
#将要分析的数据导入dataframe
import pandas as pd
pd.options.display.max_rows = 10 #make display smaller
unames = ['user_id','gender','age','occupation','zip']
users = pd.read_table('F:/movielens/users.dat',sep='::',header=None,names=unames)
rnames = ['user_id','movie_id','rating','timestamp']
ratings = pd.read_table('F:/movielens/ratings.dat',sep='::',header=None,names=rnames)
mnames = ['movie_id','title','genres']
movies = pd.read_table('F:/movielens/movies.dat',sep='::',header=None,names=mnames)
```

```python
#将user,movie,rating数据融合
import pandas as pd
data = pd.merge((pd.merge(users,ratings)),movies)
data
```

```python
#根据gender求各个movie的平均分
import pandas as pd
mean_ratings = data.pivot_table('rating',index='title',columns='gender',aggfunc='mean') #根据column求mean
mean_ratings
```

```python
#当一个movie有250人评分时，才有求评分的意义，根据男，女求有意义的评分
rating_by_title = data.groupby('title').size()
active_index = rating_by_title.index[rating_by_title >= 250]
mean_ratings.loc[active_index] #注意要加.loc[]，否则可能出现辨认不出的情况
```

```python
#查看女性评分最高的top10电影
top_female_ratings = mean_ratings.sort_values(by='F',ascending=False)
top_female_ratings[:10]
```

```python
#找出男，女喜好最不相同的top5电影
mean_ratings['diff'] = mean_ratings['F']-mean_ratings['M']
sorted_by_diff = mean_ratings.sort_values(by='diff',ascending=False)
print('top5.f:','\n',sorted_by_diff[:5])#找出差异最大，女生最喜欢的电影top5
print('\n','top5.m:','\n',sorted_by_diff[::-1][:5])#找出差异最大，男生最喜欢的电影top5
```

```python
#找出个人喜好差异最大的top5电影
rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title.loc[active_index].sort_values(ascending=False)[:5]
```
